<h2>Forecast for Dinner, Aug 25-28</h2>

<p>Our data shows that as time passes the likelihood of Mark buying dinner increases incredibly.</p>

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from scipy import stats

%matplotlib inline
```

<h3>Likelihood of Mark Buying Dinner</h3>

<p>Chances are in Gurpreet's favor. Looks statistically significant!</p>

```python
x = [[2],[4],[6],[10],[12],[14],[16]]
y = [0.50, 0.48, 0.60, 0.70, 0.75, 0.80, 0.75]

regr = linear_model.LinearRegression()
regr.fit(x, y)

plt.scatter(x, y, color='red')
plt.plot(x, regr.predict(x), color='blue', linewidth=2)
plt.xlabel("Hours Since Mark's Email Sent")
plt.ylabel('Likelihood of Mark Buying Dinner')
plt.show()
```

<h3>Did someone say cake?</h3>
