```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
```

```python
loansData = pd.read_csv("loansData.csv")
print(loansData)
```

```python
type(loansData)
```

```python
loansData.dropna(inplace=True)
print(loansData)
```

```python
loansData.boxplot(column="Amount.Requested")
plt.show()
```

```python
loansData.hist(column="Amount.Requested")
plt.show()
```

```python
stats.probplot(loansData["Amount.Requested"], dist="norm", plot=plt)
plt.show()
```

```python
stats.probplot(loansData["Amount.Requested"], dist="uniform", plot=plt)
plt.show()
```

```python
# Notes: 
# Median amount requested is ~$50,000 
# Range is less than ~$1,000 to ~$115,000
# Interquartile range is between ~$25,000 to ~$78,000
# There are no outliers 
# The number of requests is pretty evenly distributed for each amount requested (~250 requests)
# Amount requested is a uniform distribution
```
