<h3>Cleaning the data</h3>

This is where I start wrangling with my data. Again, I'm interested in determining whether loan length, loan amount and FICO scores determine interest rate. 

I first print the first 5 rows of each column to take a closer look at the format of my data.

```python
# first 5 rows of Interest.Rate
int_rate = loansData["Interest.Rate"][0:5]
print(int_rate)

# first 5 rows of Loan.Length
loan_len = loansData["Loan.Length"][0:5]
print(loan_len)

# first 5 rows of FICO.Range
fico_range = loansData["FICO.Range"][0:5]
print(fico_range)
```

I remove the % from interest rate and use floating values.

```python
# rstrip removes character at the end of a string 
# float converts to floating point 
# round the number to 4 digits

clean_ir = loansData["Interest.Rate"].map(lambda x: round(float(x.rstrip("%"))/100, 4))
loansData["Interest.Rate"] = clean_ir
```

I remove strings from the loan length column.

```python
# remove months from loans length

clean_loans_length = loansData["Loan.Length"].map(lambda x: float(x.rstrip("months")))
loansData["Loan.Length"] = clean_loans_length
```

FICO scores are displayed in ranges and need to be turned into numerical values. I make the range into a list with the lower range and the upper range. I then use the lower number as my FICO score. Since the range is small, this shouldn't be a problem.

```python
# convert FICO score into numerical value 

# split the range by hyphen 
clean_fico = loansData["FICO.Range"].map(lambda x: x.split("-"))

# make values into a list, only select the first value
clean_fico = clean_fico.map(lambda x: [int(n) for n in x][0])

# change the current values of the column to new variable 
loansData["FICO.Range"] = clean_fico

# rename FICO.Range to FICO.Score
loansData.rename(columns={"FICO.Range": "FICO.Score"}, inplace=True)
print(loansData)
```

<h3>Linear Regression</h3>

```python
intrate = loansData["Interest.Rate"]
loanamt = loansData["Amount.Requested"]
fico = loansData["FICO.Score"]
pd.DataFrame({"intrate": intrate, "loanamt": loanamt, "fico": fico})
```

```python
# reshape the data 
y = np.matrix(intrate).transpose()
x1 = np.matrix(fico).transpose()
x2 = np.matrix(loanamt).transpose()
```

```python
# make fico rate and loan amount into a matrix
x = np.column_stack([x1, x2])
print(x)
```

```python
# create a linear model 
X = sm.add_constant(x)
model = sm.OLS(y,X)
f = model.fit()
```

Low p value and high R squared value show that FICO score and loan amount are correlated with interest rates. R squared measures how close data is to the fitted regression line. P value indicates taht there is a statistically significant relationship between the variables and response. 

P value is 0, so null hypothesis can be rejected. R squared value is 65.7%.

Details of OLS Regression rsults (taken from (http://www.datarobot.com/blog/ordinary-least-squares-in-python/))

<strong>R-squared</strong>:	The coefficient of determination. A statistical measure of how well the regression line approximates the real data points

<strong>Adj. R-squared</strong>:	The above value adjusted based on the number of observations and the degrees-of-freedom of the residuals

<strong>F-statistic</strong>:	A measure how significant the fit is. The mean squared error of the model divided by the mean squared error of the residuals

<strong>Prob (F-statistic)</strong>:	The probability that you would get the above statistic, given the null hypothesis that they are unrelated

<strong>Log-likelihood</strong>:	The log of the likelihood function.

<strong>AIC	The Akaike Information Criterion</strong>: Adjusts the log-likelihood based on the number of observations and the complexity of the model.

<strong>BIC	The Bayesian Information Criterion</strong>: Similar to the AIC, but has a higher penalty for models with more parameters.


<strong>coef</strong>:	The estimated value of the coefficient

<strong>std err</strong>:	The basic standard error of the estimate of the coefficient. More sophisticated errors are also available.

<strong>t </strong>:	The t-statistic value. This is a measure of how statistically significant the coefficient is.

<strong>P > |t|</strong>:	P-value that the null-hypothesis that the coefficient = 0 is true. If it is less than the confidence level, often 0.05, it indicates that there is a statistically significant relationship between the term and the response.

<strong>[95.0% Conf. Interval]</strong>:	The lower and upper values of the 95% confidence interval

<strong>Skewness</strong>:	A measure of the symmetry of the data about the mean. Normally-distributed errors should be symmetrically distributed about the mean (equal amounts above and below the line).

<strong>Kurtosis</strong>:	A measure of the shape of the distribution. Compares the amount of data close to the mean with those far away from the mean (in the tails).
Omnibus	Dâ€™Angostinoâ€™s test. It provides a combined statistical test for the presence of skewness and kurtosis.

<strong>Prob(Omnibus)</strong>:	The above statistic turned into a probability

<strong>Jarque-Bera</strong>:	A different test of the skewness and kurtosis

<strong>Prob (JB)</strong>:	The above statistic turned into a probability

<strong>Durbin-Watson</strong>:	A test for the presence of autocorrelation (that the errors are not independent.) Often important in time-series analysis

<strong>Cond. No</strong>:	A test for multicollinearity (if in a fit with multiple parameters, the parameters are related with each other).

<h3>Logistic Regression</h3>

For this logistic regression, we are concerned about the odds of getting a loan for $10,000 at an interest rate less than 12% with a FICO score of 750. 

We'll use a logit function (log odds function). A probability (p) of more than 70% means we will get a loan. 

We plug in our linear equation into our logistic function. 

Y (dependent variable) = Yes/No 

X (independent variables) = Loan Amount & FICO Score

Our function is: log(p/1-p) = mx + b


<h4>Data Cleaning</h4>

```python
# check if interest rate is less than 12%; boolean 
IR_TF = intrate < .12

# apply to binary (0, 1) values  
IR_TF = IR_TF.astype(int)

# create new column 
loansData["IR_TF"] = IR_TF

# check first 10 rows
loansData.head(10)
```

```python
# Statsmodel needs an intercept column in the dataframe, so add an intercept column with constant intercept of 1.0
loansData["Intercept"] = 1.0

# check new column added
loansData.head()
```

```python
# create list of column names of independent variables 
ind_vars = loansData[["FICO.Score", "Amount.Requested"]]
ind_vars
```

Our linear model shows that interest rates varies depending on FICO score and loan amount requested. 

```python
# interest_rate = a1(FICOScore) + a2(LoanAmount) + b 
# Example: interest_rate = a1(750) + a2(10000) + b 
```

Let's define our logistic regression model.

```python
logit = sm.Logit(IR_TF, ind_vars)
```

Fit the model

```python
result = logit.fit()
```

Let's get our fitted cofficients from our results. These are coefficients of each independent variable.

```python
coeff = result.params 
print(coeff)
```
